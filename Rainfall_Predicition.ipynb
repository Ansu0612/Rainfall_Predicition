```python
# Importing Libraries

import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/austin_final.csv")
df
```

```python
X = df.drop(['PrecipitationSumInches'], axis = 1)
```

```python
Y = df['PrecipitationSumInches']
```

```python
# Reshaping it into 2D Vector

Y = Y.values.reshape(-1,1)
```

```python
Y
```

```python
day_index = 798
days = [i for i in range(Y.size)]
```

```python
# initialize the linear regression classifier

clf = LinearRegression()

# train the classifier

clf.fit(X,Y)
```

Plot the graphs:-

```python
print("The Precipitation Trend Graph: ")
plt.scatter(days, Y, color = "blue")
plt.scatter(days[day_index], Y[day_index], color = "red")
plt.title("Precipition Level")
plt.xlabel("days")
plt.ylabel("Precipition in Inches")
plt.show()
X_vis = X.filter(['TempAvgF','DewPointAvgF','HumidityAvgPercent','VisibilityAvgMiles','SeaLevelPressureAvgInches','WindAvgMPH	'])
```

```python
print("The Precepitation vs attributes trend graph:- ")

for i in range (X_vis.columns.size):
  plt.subplot(3,2,i+1)
  plt.scatter(days,X_vis[X_vis.columns.values[i][:100]], color = 'green')
  plt.scatter(days[day_index],X_vis[X_vis.columns.values[i]][day_index], color = 'red')
  plt.title(X_vis.columns.values[i])
plt.show()
```
